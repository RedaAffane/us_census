```python
census = pd.read_csv("/home/reda/Documents/dataiku_test_data/us_census_full/census_clean.csv",index_col=0)
```

```python
from sklearn import preprocessing
census.drop(["migration code-change in msa","migration code-change in reg","migration code-move within reg","migration prev res in sunbelt",'instance weigth'],inplace=True,axis=1)
new_df = census.select_dtypes(include=["object"]).apply(preprocessing.LabelEncoder().fit_transform)
census_num = census.drop(new_df.columns.values,axis=1)
census_num = pd.concat([census_num,new_df],axis=1)
census_num.head(1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier

from sklearn.metrics import recall_score, precision_recall_curve, roc_curve, roc_auc_score

X=census_num.ix[:, census_num.columns != "income"].values
y=census_num["income"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,stratify=y ,random_state=42)
```

```python
def compute_metrics(model,X_test,y_test):
    metrics={}
    metrics["precision"], metrics["recall"], _ = precision_recall_curve(y_test,model.predict_proba(X_test)[:,1])
    metrics["fpr"], metrics["tpr"], _ = roc_curve(y_test, model.predict(X_test))
    metrics["accuracy"] = model.score(X_test,y_test)
    metrics["recalls"] = recall_score(model.predict(X_test),y_test)
    metrics["auc"] = roc_auc_score(y_test,model.predict_proba(X_test)[:,1])
    return metrics
```

```python
def plot_model(models,X_train,X_test,y_train,y_test):
    plt.figure(figsize=(10,10))
    accuracy={}
    recall={}
    auc={}
    for model_name, model in models.items():
        model_fit = model.fit(X_train,y_train)
        metrics = compute_metrics(model_fit,X_test,y_test)
        
        plt.subplot(2,1,1)
        plt.plot(metrics["recall"],metrics["precision"],label=str(model_name))
        plt.title('recall-precision curve')
        plt.xlabel('recall')
        plt.ylabel('precision')
        plt.legend(loc="lower right")

        plt.subplot(2,1,2)
        plt.plot(metrics["fpr"],metrics["tpr"],label=str(model_name))
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.title('ROC curve')
        plt.xlabel('false positive rate')
        plt.ylabel('false negative rate')
        plt.legend(loc="lower right")

        accuracy[model_name] = metrics["accuracy"]
        recall[model_name] = metrics["recalls"]
        auc[model_name] = metrics["auc"]
    plt.figure()    
    sns.barplot(x=list(accuracy.keys()), y=list(accuracy.values()),palette="Blues_d")
    xt = plt.xticks(rotation=45)
    print("accuracy:\n")
    print(accuracy) 
    print("recall:\n")
    print(recall)
    print("AUC:\n") 
    print(auc) 
```

```python
plot_model({"Logistic Regression":LogisticRegression(),"Logistic Regression balanced weights":LogisticRegression(class_weight={1: 10}),"Random Forest":RandomForestClassifier(n_estimators=100),"AdaBoost":AdaBoostClassifier(n_estimators=100)},X_train, X_test, y_train, y_test)
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE

#Oversampling
ros = RandomOverSampler()
X_train_oversampled, y_train_oversampled = ros.fit_sample(X_train, y_train)

#SMOT
sm = SMOTE(kind='regular')
X_train_smot, y_train_smot = sm.fit_sample(X, y)
```

```python
plot_model({"Logistic Regression":LogisticRegression(),"Logistic Regression balanced weights":LogisticRegression(class_weight={1: 10}),"Random Forest":RandomForestClassifier(n_estimators=100),"AdaBoost":AdaBoostClassifier(n_estimators=100)},X_train_oversampled, X_test, y_train_oversampled, y_test)
```

```python
plot_model({"Logistic Regression":LogisticRegression(),"Logistic Regression balanced weights":LogisticRegression(class_weight={1: 10}),"Random Forest":RandomForestClassifier(n_estimators=100),"AdaBoost":AdaBoostClassifier(n_estimators=100)},X_train_smot, X_test, y_train_smot, y_test)
```
